```python
plt.title('Histogram:Price')
plt.xlim(3000,50000)
plt.ylim(0,70)
plt.xlabel('Price of cars')
plt.ylabel('Frequency')
sns.histplot(data=df, x='price',color='green');
```

```python
# creating a boxplot with seaborn
sns.boxplot(data=df, x='curb_weight', color='purple');
```
